There are columns we need to fix:
- Handle date

##1-0 Handle date

```python
d12['order payment date'] = pd.to_datetime(d12['order payment date'])
d12.info()
```

##1-1 Drop Col

```python
#Drop Unnecessary Columns

d12.drop(columns=['二级商品行业｜lvl2_product_industry', 'Total_Items', 'Orders', 'global seller id'], inplace=True)
```

##1-2 Sumup

```python
#group sum
numeric_col = ['GMV', 'quasi_settlement_amt_usd',
       'Longterm_Settlement_GMV', 'GMV_ShopTab', 'GMV Affiliate']

cat_col = ['order payment date', 'product id', 'product name', 'shop code',
       'shop name', 'First category name', 'Second Cat (By Team)',
       'third category name', '一级商品行业｜lvl1_product_industry', 'content_type',
       'seller am email', 'Avg_Item_Price']

d12 = d12.groupby(cat_col)[numeric_col].sum()
```

```python
d12.head()
```

```python
#I dont want that pivot format, so need to reset index

d12.reset_index(inplace=True)
```

```python
d12.head()
```

##1-3 Transpose gmv_content

```python
#I dont want content_type as rows, I want to transpose  GMV by content_type to columns

#First, create a df with no content type, then will join later
cat_col_no_contenttype =['order payment date', 'product id', 'product name', 'shop code',
       'shop name', 'First category name', 'Second Cat (By Team)',
       'third category name', '一级商品行业｜lvl1_product_industry',
       'seller am email', 'Avg_Item_Price']
```

```python
d12_nocontent = d12.groupby(cat_col_no_contenttype)[numeric_col].sum()
d12_nocontent.reset_index(inplace=True)
d12_nocontent.head()
```

```python
#Now calculate gmv by content type, and transpose

d12_content = d12.groupby(cat_col)['GMV'].sum()
d12_content = d12_content.reset_index()
d12_content.head()
```

```python
#Now I want to transpose content_type to columns

d12_content = d12_content.pivot(index=cat_col_no_contenttype, columns='content_type', values='GMV')
d12_content.head()
```

```python
d12_content.reset_index(inplace=True)
d12_content.head()
```

```python
#Now impute all NaN with 0

d12_content.fillna(0, inplace=True)
d12_content.head()
```

```python
#Now I will merge d12_nocontent and d12_content, take only live, product_card, video from d12_content

x = pd.merge(d12_nocontent, d12_content, on=cat_col_no_contenttype, how='left')
x.head()
```

```python
x.info()
```

```python
d12.info()
```

```python
d12 = x.copy()
```
